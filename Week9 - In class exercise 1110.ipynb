    a. In a jupyter notebook markdown cell, describe the difference between linear and rank correlation.
    
   *Linear correlations* describe a variable (x) and a variable (y) that can be meaningfully and accurately be applied to a linear function. For a linear model, you also need to assume a linear relationship between x and y, homoscedasticity (residuals about the same for each value of x), independence of observations, and normality (Y is normally distributed for any value of x).
   
   *Rank correlations* describe when the order between a variable (x) and a variable (y) is meaningful and important. Because of this, the data should be ordinal for a rank correlation. Rank correlation is calculated similarly to linear correlations. The primary difference is that this method uses the ranks for calcuations, as opposed to the actual values.        

    b. Create a scatter plot showing the relationship between calculated_host_listings_count and availability_365 in the Belize Airbnb data

```python
# Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# File to load
belize = "belize_airbnb_listings.csv"

# Read file into Pandas data frame
belize_df = pd.read_csv(belize)
```

```python
from textwrap3 import wrap

# Build the scatter plot
sns.set_style("whitegrid")

g = sns.scatterplot(data=belize_df,
                    x='availability_365',
                    y='calculated_host_listings_count',
                    alpha=0.9,
                    edgecolor='grey',
                    hue='calculated_host_listings_count',
                    linewidth=0.8,
                    legend=False)

# Add plot title and adjust the y-axis label to include units
#g.set_title('There is a weak correlation between the number of listings and days available', size=12, weight='bold', pad=15)
g.set_ylabel('Host listings count', size=11)
g.set_xlabel('Days available throughout the year (max=365 days)', size=11)  

title = 'There is a weak correlation between the number of listings a host has and the number of days an air bnb is available'
# Display plot
plt.title('\n'.join(wrap(title,60)))
plt.show()
```

    c. Calculate the correlation coefficient for the data in the scatter plot, and describe what is happening.

```python
# Calculate the correlation between availability and listings count
belize_df['availability_365'].corr(belize_df['calculated_host_listings_count'])
```

There is a weak, positive association between the number of days that an airbnb is available and the number of listings a host has. However, because this correlation is so weak, we cannot meaningfully predict the host listing count based on the number of days a bnb is available.

**2. Create a box and whisker plot for the Belize Airbnb data price.** 

In a markdown cell, describe what is going on in as simple of terms as possible. Use the documentation https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.boxplot.html

```python
# Describe price variable to get a sense of distribution
belize_df['price'].describe()
```

The box plots show us that there is wide variability in our distribution of nightly prices of air bnb rentals and the data are slightly skewed right. There is a range of about $\$$20 per night as the minimum with a maximum price of about $\$$1200

When we remove outliers, we can see that the median price is a little over $\$$200 and the inner quartile range is close to $\$$250 per night.

**With Outliers**

```python
# See if the distribution changes at all when looking by region
g2 = sns.boxplot(x='price',
           palette="Blues",
                data=belize_df)

g2.set_title('Distribution of Airbnb Prices', size=12, weight='bold')
g2.set_xlabel('Price per night')
g2.set_ylabel('')

# Save figure and then display
plt.show()
```

**Without Outliers**

```python
# See if the distribution changes at all when looking by region
g2 = sns.boxplot(x='price',
           palette="Blues", showfliers=False,
                data=belize_df)

g2.set_title('Distribution of Airbnb Prices', size=12, weight='bold')
g2.set_xlabel('Price per night')
g2.set_ylabel('')

# Save figure and then display
plt.show()
```
