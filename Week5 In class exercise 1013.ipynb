1. Make a function that can take any non-negative integer as an argument and return it with its digits in descending order. Essentially, rearrange the digits to create the highest possible number. If no number is passed in, make the default value inputted = 111111. Input = 43905 Output = 95430 Input = 584366 Output = 866543

```python
# Here is my function
def Integer_sort(a=111111):
    if a > 0:
        return int("".join(sorted(str(a), reverse=True)))
    else:
        print("please enter a non-negative integer")

# Sample output
print(Integer_sort(43905))
print(Integer_sort(584366))
Integer_sort(-15)
```

```python
# Import libraries
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
jeopardy_data = "JEOPARDY_CSV.csv"

# Read Purchasing File and store into Pandas data frame
jeopardy_data_df = pd.read_csv(jeopardy_data)
```

```python
# Explore data
jeopardy_data_df

# How many rows are there in the dataframe? What does a row represent?
#### This output shows that there are 216930 rows. 
#### Each row represents a question that was asked during a specific jeopardy episode
```

```python
# How many unique shows were there each year?

# First, parse out the 'year' from the Air date variable
years = pd.DatetimeIndex(jeopardy_data_df[' Air Date']).year

# Next, group this by 'years' (that we just created)
grouped_years = jeopardy_data_df.groupby(years)['Show Number']

# Find the unique per year
grouped_years.nunique()
```

```python
# Show as data frame
jeo_df = pd.DataFrame(grouped_years.nunique())
jeo_df
```

**3. Create any graph using matplotlib that helps you understand the jeopardy data. Explain what your graph shows and why it is useful.**

This graph shows the total Jeopardy shows per year. It is useful because we can quickly see how popular Jeopardy has been over time. We can also see which years were particularly successful. 

```python
import matplotlib.pyplot as plt
```

```python
# Create the bar plot
jeo_df['Show Number'].plot(kind='bar', color='b', 
                                       xlabel='',
                                       ylabel='Total Number of Shows')
plt.xticks(rotation=90)
plt.title('Total Jeopardy Shows Per Year', fontsize=12, fontweight='bold')

# Display our chart
plt.show()
```
